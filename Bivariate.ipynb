```python
from statsmodels.stats.outliers_influence import variance_inflation_factor

def calc_vif(X):

    # Calculating VIF
    vif = pd.DataFrame()
    vif["variables"] = X.columns
    vif["VIF"] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]

    return(vif)
```

# satsmodels provides for Variance inflation factor
# VIF only accepts quantitative datas
# 1st we need to create a Data Frame and store in the vif variable.
# X.column, X is the input. whatever input we give that would be stored in the "variable"
# in the 5th line of code - the for loop gets executed 1st (for i in range(X.shape[1])]. X.shape represents the number of rows and [1] represents number of Column.
# the for loop gets executed then (X.values,i) X.values finds the variance inflation factor.
# the end result will be stored from variance inflation factor to vif["VIF"].
# this is how we can calculate VIF using this function.

```python
calc_vif(dataset[["etest_p","salary"]])
```
